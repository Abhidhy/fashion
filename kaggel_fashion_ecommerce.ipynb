```python
rating_with_totalRatingCount = df1.merge(num_rating_df, left_on = 'title', right_on = 'title', how = 'left')
rating_with_totalRatingCount.head()
```

Collaborative Filtering Based Recommender System

```python
popularity_threshold = 40
rating_popular_movie= rating_with_totalRatingCount.query('num_ratings >= @popularity_threshold')
rating_popular_movie.head()
```

```python
rating_popular_movie.shape
```

```python
len(df1)
```

```python
## First lets create a Pivot matrix

item_features_df=rating_popular_movie.pivot_table(index='title',columns='category',values='num_ratings').fillna(0)
item_features_df.head()
```

```python
item_features_df.isnull().sum()
```

```python
item_features_df.shape
```

```python
from scipy.sparse import csr_matrix

item_features_df_matrix = csr_matrix(item_features_df.values)

from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(item_features_df_matrix)
```

```python
item_features_df.shape
```

```python
import numpy as np
```

```python
query_index = np.random.choice(item_features_df.shape[1])
print(query_index)
distances, indices = model_knn.kneighbors(item_features_df.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
item_features_df.head()
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(item_features_df.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, item_features_df.index[indices.flatten()[i]], distances.flatten()[i]))
```

2nd method

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(item_features_df)
```

```python
similarity_scores.shape
```

```python
def recommend(title):
    # index fetch
    index = np.where(item_features_df.index==title)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = df1[df1['title'] == item_features_df.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('title')['title'].values))
        item.extend(list(temp_df.drop_duplicates('title')['product_detials'].values))
        # item.extend(list(temp_df.drop_duplicates('title')['colour'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('Rayon Embroidered Flared Kurta')
```
